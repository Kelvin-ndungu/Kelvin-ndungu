```python
#When creating a new class one has we need to use the class keyword, name of the class, and the class parent. 

#import a library for plotting

import matplotlib.pyplot as plt
%matplotlib inline

class circle(object):
  #constructor 

  #double underscore __init__
  def __init__(self, radius = 3, color = 'blue'):
    self.radius = radius
    self.color = color

  #Method
  def add_radius(self, r):
    self.radius = self.radius + r
    return(self.radius)

  #method
  def draw_circle(self):
    plt.gca().add_patch(plt.Circle((0,0),radius = self.radius, fc= self.color))
    plt.axis("scaled")
    plt.show()

# Code above creates a class and the methods. 

#We can now create objects/ instances of the class. 

red_circle = circle(10, "red")

#we can use the dir command to get the list of the object's methods

#dir(red_circle)

# We can also look at the different data attributes

red_circle.radius
red_circle.color

#We can change the attributes
red_circle.radius = 12

#We can call the draw_circle method
red_circle.draw_circle()

#We can increase the radius by calling the add radius method
```

```python
# We can create a class for drawing rectangles 

class rectangle(object):

  def __init__(self, width = 5, height = 8, color = "red"):
    self.height = height
    self.width = width
    self.color = color
  
  #method
  def draw_rectangle(self):
    plt.gca().add_patch(plt.Rectangle((0,0),self.width, self.height, fc=self.color))
    plt.axis("scaled")
    plt.show()

#After creating the class we can draw a red rectangle. 
red_rectangle =rectangle(10, 20, 'red')
red_rectangle.draw_rectangle()

#One can change the data attributes to create new objects. 
```
